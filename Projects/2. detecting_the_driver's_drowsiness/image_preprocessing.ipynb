```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def light_removing(frame) :
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    lab = cv2.cvtColor(frame, cv2.COLOR_BGR2LAB)
    L = lab[:,:,0]
    med_L = cv2.medianBlur(L,99) #median filter
    invert_L = cv2.bitwise_not(med_L) #invert lightness
    composed = cv2.addWeighted(gray, 0.75, invert_L, 0.25, 0)
    return L, composed, gray, med_L, invert_L
```

```python
ho = cv2.imread("ho.jpg")
L, composition, gray, med_L, invert_L = light_removing(ho)

# cv2.imwrite("output.jpg", np.hstack([mart, mart_blur_adaptive_thr]))

cv2.imwrite("L.jpg",L)
# cv2.imshow("L", L)
# plt.show()

cv2.imwrite("med_L.jpg", med_L)
# plt.show()

cv2.imwrite("invert_L.jpg", invert_L)

# ho_original= cv2.cvtColor(ho, cv2.COLOR_BGR2RGB)
# cv2.imwrite("ho_original.jpg", ho_original)
# plt.show()

cv2.imwrite("composition.jpg", composition)

cv2.imwrite("Ls.jpg", np.hstack([L, med_L, invert_L]))

ho_gray = cv2.cvtColor(ho, cv2.COLOR_BGR2GRAY)
cv2.imwrite("ho_gray.jpg", ho_gray)
# plt.show()


# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
