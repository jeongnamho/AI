```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
import cv2
```

```python
train=cv2.imread('patch3.png')
b,g,r=cv2.split(train)


hr=plt.hist(r.ravel(),256,[0,256],color='r')
hg=plt.hist(g.ravel(),256,[0,256],color='g')
hb=plt.hist(b.ravel(),256,[0,256],color='b')

maxr=(np.where(hr[0]==np.max(hr[0])))[0][0]
maxg=(np.where(hg[0]==np.max(hg[0])))[0][0]
maxb=(np.where(hb[0]==np.max(hb[0])))[0][0]

print(maxr,'\n',maxg,'\n',maxb)
```

```python
test=cv2.imread('chroma key.png')
test_rgb=test.copy()
test_hsv=test.copy()
```

```python
for r in range(test.shape[0]):
    for c in range(test.shape[1]):
        if np.array_equal(test[r,c,:],[maxb,maxg,maxr]):
            test_rgb[r,c,:]=[0,0,0]
plt.figure(figsize=(30,10))            
plt.imshow(test_rgb)
```

```python
h,s,v=cv2.split(cv2.cvtColor(train,cv2.COLOR_BGR2HSV))

hh=plt.hist(h.ravel(),256,[0,256],color='r')
hs=plt.hist(s.ravel(),256,[0,256],color='g')
hv=plt.hist(v.ravel(),256,[0,256],color='b')

maxh=(np.where(hh[0]==np.max(hh[0])))[0][0]
maxs=(np.where(hs[0]==np.max(hs[0])))[0][0]
maxv=(np.where(hv[0]==np.max(hv[0])))[0][0]

for r in range(h.shape[0]):
    for c in range(h.shape[1]):
        if np.array_equal(test[r,c,:],[maxh,maxs,maxv]):
            test_hsv[r,c,:]=[0,0,0]
plt.figure(figsize=(30,10))            
plt.imshow(test_hsv)
```
