```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
import cv2
from PIL import ImageFont,ImageDraw,Image
import scipy.stats as sp
```

```python
def imshow(tit,image):
    plt.figure(figsize=(20,10))
    plt.title(tit)
    if len(image.shape)==3:
        plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
    else:
        plt.imshow(image, cmap='gray')
    plt.show()
```

```python
rec=cv2.imread('contour.jpg')
imshow('',rec)
```

```python
gray=cv2.cvtColor(rec,cv2.COLOR_BGR2GRAY)
gray_b=cv2.medianBlur(gray,37)
ath=cv2.adaptiveThreshold(gray_b,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,31,9)
ath=255-ath
imshow('',ath)

contours,_=cv2.findContours(ath,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
```

```python
contours,_=cv2.findContours(ath,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
```

```python
con=contours[0]
```

```python
peri=cv2.arcLength(con, True)
approx=cv2.approxPolyDP(con,0.02*peri,True)
```

```python
print(peri,'\n',approx)
```

```python
cont=cv2.drawContours(rec,[approx],-1,(0,0,255),10)
```

```python
imshow('',cont)
```
