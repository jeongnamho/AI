```python
import requests
import json
from pprint import pprint # 예쁘게 출력해주려고!
```

```python
def get_answer(text, sessionId):
    data_send = {
        'query': text,
        'sessionId': sessionId, # 유저구별
        'lang': 'ko',
        'timezone' : 'Asia/Seoul'
    }
    data_header = {
        'Authorization': 'Bearer 0500750d7de743e18a83d6cba940c0fd',
        'Content-Type': 'application/json; charset=utf-8'
    }

    dialogflow_url = 'https://api.dialogflow.com/v1/query?v=20150910' 
    res = requests.post(dialogflow_url, data=json.dumps(data_send), # data에는 일반적인 데이터 형태(브라우저에선 안보임), 타입은 json타입 
                        headers=data_header)   # header : 딕셔너리
    if res.status_code == requests.codes.ok:
       return res.json()   
    return {}
```
