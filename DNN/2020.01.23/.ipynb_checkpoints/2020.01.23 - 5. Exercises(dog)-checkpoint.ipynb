## 한꺼번에 하기

```python
from keras.preprocessing.image import ImageDataGenerator

test_datagen = ImageDataGenerator()

test_generator = test_datagen.flow_from_directory(
        'cnntest',
        target_size=(224,224),
        batch_size=3,
        class_mode='binary', shuffle=False)
X, T = test_generator.next()
print(X.shape[0])
print(X.shape) # (3개의 파일, 224*224 사이즈, RGB채널(3))
```

```python
decode_vgg = decode_predictions(
    vgg_model.predict(
        vgg16.preprocess_input(X.copy())), top=3)

decode_resnet = decode_predictions(
    resnet_model.predict(
        resnet50.preprocess_input(X.copy())), top=3)

decode_mobilenet = decode_predictions(
    mobilenet_model.predict(
        mobilenet.preprocess_input(X.copy())), top=3)
```

```python
for i in range(X.shape[0]):
    img = cv2.resize(X[i], (1000,1000))
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    cv2.putText(img, "VGG16: {}".format(decode_vgg[i][0][1]) , (50,50), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 3)
    cv2.putText(img, "rsnet: {}".format(decode_resnet[i][0][1]) , (50,100), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 3)
    cv2.putText(img, "mobilenet: {}".format(decode_mobilenet[i][0][1]) , (50,150), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 3)
    cv2.imwrite(f"dog{i+1}.jpg",img)
```

### 하나씩 하기

```python
filename = 'dog.jpg'
original = load_img(filename, target_size=(224, 224)) # 위의 summary에서 보면 input 영상이 224*224이므로 
                                                      # 로딩할때 사이즈를 맞춰준다.

numpy_image = img_to_array(original)
numpy_image = np.uint8(numpy_image)

image_batch = np.expand_dims(numpy_image, axis=0) # 이미지 reshape
image_batch = np.uint8(image_batch)
```

```python
fig=plt.figure(figsize=(15, 5))
columns = 3
rows = 1
imgs = [original, numpy_image, image_batch[0]]
names = ['original', 'numpy_image', 'image_batch']
sizes = ["none", numpy_image.shape, image_batch.shape]

for i in range(columns*rows):
    img = imgs[i]
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(img)
    plt.title(f"{names[i]}, {sizes[i]}")
plt.show()
```

```python
processed_image = vgg16.preprocess_input(image_batch.copy())
predictions = vgg_model.predict(processed_image)
label_vgg = decode_predictions(predictions, top=10) 

processed_image = resnet50.preprocess_input(image_batch.copy())
predictions = resnet_model.predict(processed_image)
label_resnet = decode_predictions(predictions, top=3)

processed_image = mobilenet.preprocess_input(image_batch.copy())
predictions = mobilenet_model.predict(processed_image)
label_mobilenet = decode_predictions(predictions)
```

```python
img = cv2.imread("dog.jpg")
img = cv2.resize(img, (1000,1000))
cv2.putText(img, "VGG16: {}, {:.2f}".format(label_vgg[0][0][1], label_vgg[0][0][2]) , (200, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 3)
cv2.putText(img, "MobileNet: {}, {:.2f}".format(label_mobilenet[0][0][1], label_mobilenet[0][0][2]) , (200, 75), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 3)
# cv2.putText(img, "Inception: {}, {:.2f}".format(label_inception[0][0][1], label_inception[0][0][2]) , (200, 110), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 3)
cv2.putText(img, "ResNet50: {}, {:.2f}".format(label_resnet[0][0][1], label_resnet[0][0][2]) , (200, 145), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 3)
cv2.imwrite("out2.jpg", img)
```
