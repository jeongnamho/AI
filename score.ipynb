```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rc("font", family = "Malgun Gothic")

# 글씨가 뭉게져서 보이는 것 같으면 아래의 설정을 통해서 개선할 수 있음
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
```

```python
df = pd.read_excel('score.xlsx', index_col=[0], header=[0,1])
df["언어오답률"] = df["언어", "틀린"]/df["언어", "total"]
df["수리오답률"] = df["수리","틀린"]/df["수리", "total"]
df["추리오답률"] = df["추리","틀린"]/df["추리", "total"]
df["시각오답률"] = df["시각","틀린"]/df["시각", "total"]
df
```

```python
hackers = df.loc[df["Unnamed: 1_level_0", "교재"] == "ㄱ"]
hackers_incorrect = hackers.iloc[:,[1,3,5,7]]
hackers_incorrect

hackers_incorrect_total = hackers.iloc[:,[2,4,6,8]]
hackers_incorrect_total


weport = df.loc[df["Unnamed: 1_level_0", "교재"] == "ㄴ"]
weport_incorrect = weport.iloc[:,[1,3,5,7]]
weport_incorrect

weport_incorrect_total = weport.iloc[:,[2,4,6,8]]
weport_incorrect_total




# 각 과목당 오답률
# df.iloc[:,9:]
```

```python
figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
figure.set_size_inches(15,4)

hackers_incorrect.plot(kind="bar", legend=True, ax=ax1, title="틀린개수")
hackers_incorrect_total.plot(kind="bar", legend=True, ax=ax2, title="틀린개수+못푼개수")
```

```python
figure, ([ax1, ax2], [ax3, ax4]) = plt.subplots(nrows=2, ncols=2)
figure.set_size_inches(18,12)


hackers_incorrect.plot(kind="bar", legend=True, ax=ax1, title="해커스1_틀린개수", cmap="viridis")
hackers_incorrect_total.plot(kind="bar", legend=True, ax=ax2, title="해커스1_틀린개수+못푼개수")
weport_incorrect.plot(kind="bar", legend=True, ax=ax3, title="위포트2_틀린개수")
weport_incorrect_total.plot(kind="bar", legend=True, ax=ax4, title="위포트2_틀린개수+못푼개수")
```
